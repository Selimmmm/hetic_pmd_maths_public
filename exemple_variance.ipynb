## Example variance avec loi de Poisson

## La loi de Poisson est une distribution pour laquelle la moyenne est égale à la variance

```python
import seaborn as sns
from scipy.stats import poisson
import numpy as np
# Here k = 3 and lambda = 5
poisson.pmf(3, 5)
sns.barplot(x=np.arange(13), y=poisson.pmf(np.arange(13),5), color='r')
plt.title("Un example de loi de Poisson", fontsize=20)
plt.show()
```

## $ \mathbb{E}(X) = \mathbb{V}(X) = 1$

```python
mu = 1
mean, var, skew, kurt = poisson.stats(mu, moments='mvsk')
print(f"La variance de X est {mu}")

from scipy.stats import poisson
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)


x = np.arange(poisson.ppf(0.01, mu),
              poisson.ppf(0.99, mu))

ax.plot(x, poisson.pmf(x, mu), 'bo', ms=8, label='poisson pmf')
ax.vlines(x, 0, poisson.pmf(x, mu), colors='b', lw=5, alpha=0.5)
plt.show()
```

## Exemple de la poste à paris et à vesoul

```python


from scipy.stats import poisson
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 4, figsize=(20,10))


for i, mu in enumerate([1, 10, 100, 1000]):
    mean, var, skew, kurt = poisson.stats(mu, moments='mvsk')
    print(f"La variance de X est {mu}")

    x = np.arange(poisson.ppf(0.01, mu),
                  poisson.ppf(0.99, mu))

    ax[i].plot(x, poisson.pmf(x, mu), 'bo', ms=8)
    ax[i].vlines(x, 0, poisson.pmf(x, mu), colors='b', lw=5, alpha=0.5)
    ax[i].axvline(mu, color='red')
    ax[i].title.set_text(f"La variance et la moyenne sont de {var}")

plt.show()
```
